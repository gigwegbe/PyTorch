```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("data/data.csv",dtype=np.float64,delimiter=",")
```

```python
data.shape
```

```python
data[:5,::]
```

```python
X = data[::,0:1]
Y = data[::,-1:]
```

```python
# our model for the forward pass
def forward(x):
    return x * w
```

```python
# Loss function
def loss(x,y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y)
```

```python
w_list = []
mse_list = []

for w in np.arange(0.0, 3.1, 0.1):
    print("w=",w)
    l_sum = 0
    for x_val, y_val in zip(X,Y):
        y_pred_val = forward(x_val)
        l = loss(x_val, y_val)
        l_sum += l
        print("\t", x_val, y_val, y_pred_val, l)
    print("MSE=", l_sum /3)
    w_list.append(w)
    mse_list.append(l_sum / 3)
    
```

```python
plt.plot(w_list, mse_list)
plt.ylabel("Loss")
plt.xlabel("W")
plt.show()
```
